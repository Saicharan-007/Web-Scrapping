```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
url='https://www.imdb.com/search/title/?title_type=feature'
request=requests.get(url)
url_dt=BeautifulSoup(request.text,'lxml')
url_tg=url_dt.find_all('a',attrs={'class': None})
url_tg_dt=[tag.attrs['href'] for tag in url_tg if tag.attrs['href'].startswith('/title') and tag.attrs['href'].endswith('/')]
url_tg_dt=list(dict.fromkeys(url_tg_dt))
url_link=["https://www.imdb.com"+tag+'reviews' for tag in url_tg_dt]
print(url_link)
print(url_link[0])
print(url_link[1])
```

```python
def getindex(a):
  minposition=a.index(min(a))
  maxposition=a.index(max(a))
  return minposition,maxposition
com=[]
votes=[]
vote1=[]
uid=[]
cinema=[]
review_count=[]
for i in range(0,50):
  url_dt=BeautifulSoup(requests.get(url_link[i]).text,'html.parser')
  voting=[tag.previous_element for tag in url_dt.findAll('span',{"class":'point-scale'})]
  if len(voting)>0:
    n_index,p_index=getindex(list(map(int,voting)))
    review_list=url_dt.findAll('a',attrs={'class':'title'})
    revw=["https://www.imdb.com"+review['href'] for review in review_list]
  else:
    print(None)
  for j in revw:
    txt=BeautifulSoup(requests.get(j).text,'html.parser').find('div',attrs={'class':'text show-more__control'}).get_text()
    rating=BeautifulSoup(requests.get(j).text,'html.parser').find('div',attrs={'class':'lister-item-content'}).get_text()
    users=BeautifulSoup(requests.get(j).text,'html.parser').find('title').get_text()
    descr=BeautifulSoup(requests.get(j).text,'html.parser').find('a',attrs={'class':'title'}).get_text()
    len_rating1=len(vote1)
    if(txt==''):
      com.append('No Review')
    else:
      com.append('txt')
    votes.append(rating)
    uid.append(users.split("'s")[0])
    cinema.append(users.split('Review of')[1])
    review_count.append(descr)
  print('vote:'+str(len(votes))+' review list:'+str(len(com)))
  if (len_rating1==0):
    k=0
  else:
    k=len_rating1
  for j in range(k,len(votes)):
    vote1.append(votes[j].split('/10')[0].split('\n')[len(votes[j].split('/10')[0].split('\n'))-1])
  print('rating1:'+str(len(vote1)))
data_review={
    'Cinema Name':cinema,
    'User Name':users,
    'Review Note':review_count,
    'Detail Review':com,
    'Rating':vote1
}
df=pd.DataFrame(data_review)
df




```
